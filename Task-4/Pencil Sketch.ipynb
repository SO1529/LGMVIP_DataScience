```python
!pip install google-colab
import cv2

def display_image(window_name, image):
    """Utility function to display an image and wait for a key press."""
    cv2_imshow(image)
    cv2.waitKey(0)

image = cv2.imread('/content/download.jpeg')
b_and_w_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
inverted_image = 255 - b_and_w_image
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(b_and_w_image, inverted_blurred, scale=256.0)
display_image("Original Image", image)
display_image("Pencil Sketch Image", pencil_sketch)
cv2.destroyAllWindows()
```
